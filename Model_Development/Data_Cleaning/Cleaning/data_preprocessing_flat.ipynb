# Univariate Analysis

```python
# Society
df['society'].value_counts()
```

```python
df['society'].value_counts().shape  # Checking number of unique socities
```

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
# Removed star and rating from society
df['society'].value_counts()
```

```python
# Price
df['price'].value_counts()
```

```python
df['price'].isnull().sum()
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
# Create a function to convert all prices in same unit

def normalize_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100, 2)
        else:
            return round(float(x[0]), 2)
```

```python
df['price'] = df['price'].str.split(' ').apply(normalize_price)
```

```python
df.head()
```

```python
# price per sqft
df['price_per_sqft'].value_counts()
```

```python
# get only price per sqft
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹', '').str.replace(',','').str.strip().astype('float')
```
